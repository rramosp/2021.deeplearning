## Context 

The aim of this LAB is to test the capacity of recurrent networks to predict future values of pollution based on past vallues of pollution and other environmental time series.

Load data:

```python
import pandas as pd

# Load data (don't use parse_dates for multiple columns)
dataset = pd.read_csv("local/data/raw.csv")

# Combine year, month, day, hour into a datetime column
dataset["date"] = pd.to_datetime(
    dataset[["year", "month", "day", "hour"]],
    format="%Y %m %d %H"
)

# Set datetime as index
dataset = dataset.set_index("date")

# Drop unnecessary columns
if "No" in dataset.columns:
    dataset = dataset.drop(columns=["No", "year", "month", "day", "hour"])
else:
    dataset = dataset.drop(columns=["year", "month", "day", "hour"])

# Rename columns
dataset.columns = ["pollution", "dew", "temp", "press", "wnd_dir", "wnd_spd", "snow", "rain"]

# Fill missing pollution values with 0
dataset["pollution"] = dataset["pollution"].fillna(0)

# Drop the first 24 hours
dataset = dataset.iloc[24:]

# Summarize first 5 rows
print(dataset.head(5))

# Save compressed CSV
dataset.to_csv("local/data/pollution.csv", compression="gzip")
```
